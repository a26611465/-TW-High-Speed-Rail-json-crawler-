```python
import pandas as pd
import requests
import json

url = 'https://www.thsrc.com.tw/TimeTable/Search'

form_data = {
    'SearchType': 'S',
    'Lang': 'TW',
    'StartStation': 'TaiPei',
    'EndStation': 'JiaYi',
    'OutWardSearchDate': '2021/05/10',
    'OutWardSearchTime': '05:30',
    'ReturnSearchDate': '2021/04/26',
    'ReturnSearchTime': '18:30',
    'DiscountType': '',
}

res1 = requests.post(url, data=form_data)

#解碼
res1.content.decode('utf-8')

# 用json解析
data = json.loads(res1.text)


# 取出JSON需要的資料層
trainItem = data['data']['DepartureTable']['TrainItem']

print(trainItem)

# 所有班車(train_number)
train_number = []
for item in trainItem:
    train_number.append(item['TrainNumber'])


    
# 所有出發時間(departure_time)
departure_time = []
for item in trainItem:
    departure_time.append(item['DepartureTime'])


# 所有到達時間(arrival_time)
arrival_time = []
for item in trainItem:
    arrival_time.append(item['DestinationTime'])

# 所有行車時間(duration)
duration = []
for item in trainItem:
    duration.append(item['Duration'])



# 整理成表格
highway_df = pd.DataFrame({
    '車次': train_number,
    '出發時間': departure_time,
    '到達時間': arrival_time,
    '行車時間': duration},
    columns = ['車次', '出發時間', '到達時間', '行車時間'])

highway_df

```
